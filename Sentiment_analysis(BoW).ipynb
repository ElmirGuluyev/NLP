First, we are importing all the libraries that we need.

```python
# Libraries
import nltk
import json 
import gzip
import pandas as pd
from nltk.collocations import BigramCollocationFinder, BigramAssocMeasures
from nltk.corpus import stopwords
from collections import Counter
from nltk.tokenize import sent_tokenize, word_tokenize, TweetTokenizer
from nltk.stem import WordNetLemmatizer
from nltk import FreqDist
from nltk.util import ngrams
from nltk.collocations import *
import re
```

```python
# Extracting the list of dictionaries from the json file

data = []

with open("C:\\Users\\Lenovo\\OneDrive\\Desktop\\Amazon_Instant_Video_5.json") as f:
    for line in f:
        data.append(json.loads(line))
```

```python
data[10:13]
```

```python
# Extracting the reviews from a dictionary to a list
list_rt = [d['reviewText'] for d in data] # A list of strings for each review
list_o = [d['overall'] for d in data] # A list of strings for each overall rating
list_s = [d['summary'] for d in data] # A list of strings for each overall summary
```

```python
# Creating a txt file for the reviews and then dumping reviews into that text file
# open file in write mode
with open(r'C:\\Users\\Lenovo\\OneDrive\\Desktop\\reviewText.txt', 'w') as fp:
    for item in list_rt:
        # write each item on a new line
        fp.write("%s\n" % item)
    print('Done')
```

```python
# Creating a txt file for the overall and then dumping str into that text file
# open file in write mode
with open(r'C:\\Users\\Lenovo\\OneDrive\\Desktop\\overall.txt', 'w') as fp:
    for item in list_o:
        # write each item on a new line
        fp.write("%s\n" % item)
    print('Done')
```

```python
# Creating a txt file for the summary and then dumping str into that text file
# open file in write mode
with open(r'C:\\Users\\Lenovo\\OneDrive\\Desktop\\summary.txt', 'w') as fp:
    for item in list_s:
        # write each item on a new line
        fp.write("%s\n" % item)
    print('Done')
```

Now, as we extracted the data from the json file into lists. We can begin to tokenize and conduct other necessary operations on the data. We first start with the reviewText.

```python
# Now we will work with reviewText solely
tokenizer = TweetTokenizer()
rt_tok  = [tokenizer.tokenize(sen) for sen in list_rt] #Creating tokens from the list of strings

```

```python
# This function basically merges all the lists into one list
rt_tokens = []
for l in rt_tok:
    for word in l:
        rt_tokens.append(word)
print(len(rt_tokens))
```

```python
# Some pre-processing
stopwords_ = set(stopwords.words('english'))
rt_tokens = [word.lower() for word in rt_tokens if len(word) > 2] # Making all the words lowercase
rt_tokens = [word for word in rt_tokens if word not in stopwords_] # removing the stop words from the list
rt_tokens = [w for w in rt_tokens if re.search("^[a-zA-Z0-9]+$",w)] # Omitting non-words

lemmatizer = WordNetLemmatizer()

rt_tokens = [lemmatizer.lemmatize(w) for w in rt_tokens] # lemmatizations of the tokens 
```

We will be working with summary now.

```python
# Now we will work with summary solely
tokenizer = TweetTokenizer()
s_tok  = [tokenizer.tokenize(sen) for sen in list_s] #Creating tokens from the list of strings
```

```python
# This function basically merges all the lists into one list
s_tokens = []
for l in s_tok:
    for word in l:
        s_tokens.append(word)
print(len(s_tokens))
```

```python
# Some pre-processing
stopwords_ = set(stopwords.words('english'))
s_tokens = [word.lower() for word in s_tokens if len(word) > 2] # Making all the words lowercase
s_tokens = [word for word in s_tokens if word not in stopwords_] # removing the stop words from the list
s_tokens = [w for w in s_tokens if re.search("^[a-zA-Z0-9]+$",w)] # Omitting non-words
lemmatizer = WordNetLemmatizer()

s_tokens = [lemmatizer.lemmatize(w) for w in s_tokens] # lemmatizations of the tokens 
```

## Analysis 

Now, as we have pre-processed the data, we can move onto the analysis part

```python
# listing top 50 words by frequency for ReviewText
r_fdist = FreqDist(rt_tokens) # This creates a list of frequencies for words
r_topkeys = r_fdist.most_common(20)

# In this for loop, "pair" refers to the word together with its frequency value
for pair in r_topkeys: 
    print (pair)
```

```python
# listing top 50 words by frequency for summary
s_fdist = FreqDist(s_tokens) # This creates a list of frequencies for words
s_topkeys = s_fdist.most_common(20)

# In this for loop, "pair" refers to the word together with its frequency value
for pair in s_topkeys: 
    print (pair)
```

```python
# Listing top 50 bigrams for ReviewText
r_bigrams = [" ".join(w) for w in ngrams(rt_tokens, 2)]
r_fdist_bi = FreqDist(r_bigrams) # This creates a list of frequencies for bigrams

r_fdist_bi.most_common(50)
```

```python
# Listing top 50 bigrams for summary
s_bigrams = [" ".join(w) for w in ngrams(s_tokens, 2)]
s_fdist_bi = FreqDist(s_bigrams) # This creates a list of frequencies for bigrams

s_fdist_bi.most_common(50)
```

```python
# listing top 50 bigrams by PMI score for ReviewText

bigram_measures = nltk.collocations.BigramAssocMeasures()
finder = BigramCollocationFinder.from_words(rt_tokens)
scored = finder.score_ngrams(bigram_measures.raw_freq)
for bscore in scored[:50]:
    print (bscore)
```

```python
# listing top 50 bigrams by PMI score for summary

bigram_measures = nltk.collocations.BigramAssocMeasures()
finder = BigramCollocationFinder.from_words(s_tokens)
scored = finder.score_ngrams(bigram_measures.raw_freq)
for bscore in scored[:50]:
    print (bscore)
```
